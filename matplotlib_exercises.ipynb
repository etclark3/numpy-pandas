```python
x = range(0,10)
y = [math.sqrt(n) for n in x]
z = [n**2 for n in x]
w = [1/(n+1) for n in x]
q = [2**n for n in x]

pp.plot(x, c='red')
pp.plot(y, c='orange')
pp.plot(z, c='purple')
pp.plot(w, c='green')
pp.plot(q, c='blue')
pp.xlabel('x')
pp.ylabel('y')
pp.title('Lot of Things')
```
