```python
iris = sns.load_dataset('iris')
iris
```

```python
iris.info()
```

```python
iris['petal_length']
```

```python
sns.relplot(x='species', y='petal_length', data=iris)
```

```python
sns.relplot(x='petal_width', y='petal_length', data=iris)
plt.title('Petal Width vs. Length')
plt.xlabel('Width')
plt.ylabel('Length')
plt.show()
# There seems to be a linear correlation
```

```python
sns.relplot(x='petal_width', y='petal_length', hue='species', data=iris)
plt.title('Petal Width vs. Length')
plt.xlabel('Width')
plt.ylabel('Length')
plt.show()
```

```python
iris
```

```python
sns.relplot(x='sepal_width', y='sepal_length', data=iris)
plt.xlim(0,5)
plt.ylim(0,9)
plt.show()
# the data plotted using sepal does not seem to show a correlation
```

```python
sns.pairplot(data=iris, hue='species')
# petal length/width appear to be the best features to use when trying to determine the species
```
