```python
x = range(1,100)
```

```python
# Q5: Make a new Jupyter notebook named big_o_notation.ipynb
# Title your chart "Big O Notation"
# Label your x axis "Elements"
# Label your y axis "Operations"
# Label your curves or make a legend for the curves
# Use LaTex notation where possible
```

```python
pp.figure(figsize=(14,8))

y1 = [0*n + 1 for n in x]
y2 = [math.log(n) for n in x if x != 0]
y3 = [n for n in x]
y4 = [n*math.log(n) for n in x if x != 0]
y5 = [pow(2,n)for n in x]
y6 = [pow(n,2)for n in x]
y7 = [math.factorial(n) for n in x]
y8 = [pow(n,n) for n in x]

pp.plot(y1, c='red', label='01')
pp.plot(y2, c='orange', label='log n')
pp.plot(y3, c='purple', label='0(n)')
pp.plot(y4, c='green', label='')
pp.plot(y5, c='blue', label='')
pp.plot(y6, c='teal', label='')
pp.plot(y7, c='gray', label='O(n!)')
pp.plot(y8, c='black', label='O(n^n)')
pp.ylim(0,20)
pp.xlim(0,20)
pp.xlabel('elements')
pp.ylabel('operations')
pp.title('Big O')
pp.suptitle('SUPER O')
pp.show()
```
